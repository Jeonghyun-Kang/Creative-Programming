### 1.1 위치 데이터 시각화
> 요금 예측에서 가장 중요한 정보는 승차 위치와 하차 위치이다. 뉴욕시 택시 요금은 대부분 운행 거리로 결정하기 때문이다.

위도경도.png

* 승하차의 위치를 뉴욕시 내로 한정한다. 
* 뉴욕시는 대략 경도 -74.05에서 -73.75사이에, 그리고 위도 40.63에서 40.85 사이에 위치한다.

```python
# 뉴욕시의 경도 범위

nyc_min_longitude = -74.05 #뉴욕시의 최소 경도(최솟값)
nyc_max_longitude = -73.75 #뉴욕시의 최대 경도(최댓값)
```

```python
# 뉴욕시의 위도 범위

nyc_min_latitude = 40.63 #뉴욕시의 최소 위도(최솟값)
nyc_max_latitude = 40.85 #뉴욕시의 최대 위도(최댓값)
```

```python
# df를 복사하여 df2를 만든다.

df2 = df.copy() #데이터프레임 df에서 위 범위(뉴욕시 안에서만 이동한 위치정보-위도X축/경도Y축)에 있는 데이터들만 필터링해서 시각화
                #df가 변경되면 안되므로 복사해서 변수df2에 대입, df2에서 뉴욕시 정보(위에 해당하는 위도/경도)만 필터링
```

* 승차위치가 뉴욕시인 데이터를 필터링한다.

```python
# 'pickup_longitude'가 nyc_min_longitude보다 크고, nyc_max_longitude는 작은 데이터들만 추출하여 df2에 대입한다.

df2 = df2.loc[(df2['pickup_longitude']>nyc_min_longitude) & (df2['pickup_longitude']<nyc_max_longitude)] #.loc[(조건식)]으로 필터링: 승차경도가 뉴욕시의 범위인 것만 걸러냄
df2                                                                                                      #조건식: (df2[컬럼명] > 최솟값) & (df2[컬럼명] < 최댓값) 최솟값보다 크고, 최댓값보다 작음
```

```python
# 'pickup_latitude'가 nyc_min_latitude보다 크고, nyc_max_latitude는 작은 데이터들만 추출하여 df2에 대입한다.

df2 = df2.loc[(df2['pickup_latitude']>nyc_min_latitude) & (df2['pickup_latitude']<nyc_max_latitude)] #.loc[(조건식)]으로 필터링: 승차위도가 뉴욕시의 범위인 것만 걸러냄
df2
```

* 하차위치가 뉴욕시인 데이터를 필터링한다.

```python
# 'dropoff_longitude'가 nyc_min_longitude보다 크고, nyc_max_longitude는 작은 데이터들만 추출하여 df2에 대입한다.

df2 = df2.loc[(df2['dropoff_longitude']>nyc_min_longitude) & (df2['dropoff_longitude']<nyc_max_longitude)] #.loc[(조건식)]으로 필터링: 하차경도가 뉴욕시의 범위인 것만 걸러냄
df2
```

```python
# 'dropoff_latitude'가 nyc_min_latitude보다 크고, nyc_max_latitude는 작은 데이터들만 추출하여 df2에 대입한다.

df2 = df2.loc[(df2['dropoff_latitude']>nyc_min_latitude) & (df2['dropoff_latitude']<nyc_max_latitude)] #.loc[(조건식)]으로 필터링: 하차위도가 뉴욕시의 범위인 것만 걸러냄
df2
```

```python
# 뉴욕시에서 승하차한 데이타만 필터링 된 df2확인

df2
```

* 뉴욕시의 핵심 랜드마트도 시각화한다.

```python
# 뉴욕시 핵심 랜트마크 데이터(경도, 위도)

landmarks = {'JFK Airport': (-73.78,40.643),
             'Laguardia Airport': (-73.87, 40.77),
             'Midtown': (-73.98, 40.76),
             'Lower Manhattan': (-74.00, 40.72),
             'Upper Manhattan': (-73.94, 40.82),
             'Brooklyn': (-73.95, 40.66)}             #뉴욕시 두개 공항과 랜드마크의 경도/위도 정보 딕셔너리{'키': (값)}로 만듦
```

```python
landmarks['JFK Airport'][1] #딕셔너리명['키'값][0번째(경도)or1번째(위도)]입력하면 데이터(값) 나옴
```

```python
# 시각화를 위해 matplotlib.pyplot임포트

import matplotlib.pyplot as plt
```

```python
# 승하차 위치를 점 그래프로 시각화하는 함수 정의

def plot_lat_long(df, landmarks, points='Pickup'): #def plot_lat_long(시각화할 데이터프레임 정보, 유명한 지역정보, points = '승/하차 정보'(디폴트는 승차로)) 그림 그리는 함수 호출
    plt.figure(figsize=(12,12))                    #함수 호출되면 사이즈 가로세로 12크기로         
    if points == 'Pickup':                         #points가 승차이면
        plt.plot(list(df['pickup_longitude']), list(df['pickup_latitude']), '.', markersize=1) #plt.plot(데이터 없음 --> 정보 다 줘야함): X축(경도), Y축(위도), 점그래프(기본 실선plot), 점의크기
    else:                                          #만약 points가 승차가 아니면                #cf)데이터프레임or시리즈(데이터 있음)에 .plot(x축 인덱스, y축 데이터)
        plt.plot(list(df['dropoff_longitude']), list(df['dropoff_latitude']), '.', markersize=1)
        
    for landmark in landmarks:                    #유명한 landamarks 정보 찍기; 딕셔너리landmarks에서 순서대로 하나씩 끄집어내서 landmark에 넣기
        plt.plot(landmarks[landmark][0], landmarks[landmark][1], '*', markersize=15, alpha=1, color='r') #plt.plot(X축: 딕셔너리명['키'][0], Y축: 딕셔너리명['키'][1], 점대신 별, 점크기, 투명도, 색깔(빨간색) )
        plt.annotate(landmark, (landmarks[landmark][0]+0.005, landmarks[landmark][1]+0.005), color='r', backgroundcolor='w') #plt.annotate(글자, 위치값(X,Y축), 글씨색깔, 배경색깔(흰색)): 글자로 꾸며주기
    
    plt.title(points + ' Locations in NYC Illustrated') #제목: 승차이면 'Pickup' + ~
    plt.xlabel('Longitude')                             #X축, Y축 label 각각 plt로 보여주기
    plt.ylabel('Latitude')
    plt.show()                                          
```

```python
# plot_lat_long()함수를 호출하여 승차위치를 점그래프로 시각화한다.

plot_lat_long(df2, landmarks, points='Pickup') #1.승차한 위치정보 #함수호출(뉴욕시의 데이터 주기, 랜드마크 주기, poitns(승차위치로)) --> x축 longitude(경도), y축 latitude(위도)
```

* 맨하탄 섬에서는 Midtown에서 가장 많이 승차한다. 그다음으로 Lower Manhattan에서 많이 승차함을 알 수 있다. 그에 비해 Upper Manhattan에서는 탑승 횟수가 적은 편이다. Upper Manhattan은 대체로 주거구역인 반면, Midtown과 LowerManhattan에는 사무실과 관광지가 몰려 있기 때문이다.
* 맨하탄 섬 바깥에서 승차하는 경우는 드물다는 것을 알 수 있다. 단, Laguardia Airport와 JFK Airport는 예외적으로 승차하는 경우가 많다.

```python
# plot_lat_long()함수를 호출하여 하차위치를 점그래프로 시각화한다.

plot_lat_long(df2, landmarks, points='Drop Off') #2.하차한 위치정보 #함수호출(뉴욕시의 데이터 주기, 랜드마크 주기, poitns(하차위치로))
```

* 승차 그래프와 비교하면, Upper Manhattan과 Brooklyn 같은 주거지역에 하차가 더 많다는 것을 알 수 있다.

### 1.2 요일 및 시간별 승차 통계
* 요일과 시간에 따른 승차 수를 집계해 본다.
* pickup_datetime 컬럼 하나에 승차 일자 및 시간 정보가 datetime 형식으로 저장되어 있다. 이 컬럼을 연, 월, 일, 요일, 시각 요소로 나누어 각각 다른 컬럼에 저장한다.

```python
# df의 첫번째 행을 확인 한다.

df.head(1) #picup_datetime컬럼: object(문자열) --> datetime(날짜시간정보)타입으로 바꿈; 연/월/일 정보 찾아볼 수 있음
```

```python
# df의 모든 컬럼의 데이터 타입을 확인한다.

df.dtypes 
```

```python
# df의 'pickup_datetime' 컬럼에서 년도를 분리하여 'year' 컬럼에 대입한다.

df['year'] = df['pickup_datetime'].dt.year #연도 정보 분리(컬럼 만들기): df['컬럼명'] = df['datetime컬럼명'].dt.year
```

```python
# df의 'pickup_datetime' 컬럼에서 달을 분리하여 'month' 컬럼에 대입한다.

df['month'] = df['pickup_datetime'].dt.month #월 정보 분리
```

```python
# df의 'pickup_datetime' 컬럼에서 일을 분리하여 'day' 컬럼에 대입한다.

df['day'] = df['pickup_datetime'].dt.day #일 정보 분리
```

```python
# df의 'pickup_datetime' 컬럼에서 요일을 분리하여 'day_of_week' 컬럼에 대입한다.

df['day_of_week'] = df['pickup_datetime'].dt.day_of_week #요일 정보 분리(0:월, 1:화, ...)
```

```python
# df의 'pickup_datetime' 컬럼에서 시간을 분리하여 'hour' 컬럼에 대입한다.

df['hour'] = df['pickup_datetime'].dt.hour #시간 정보 분리
```

#### 2.2.2 위치관련 변수

* 데이터 셋에는 승차 및 하차 위치 정보가 있지만 택시 요금에서 가장 중요한 요소인 운행 거리 정보가 없다. 따라서 <u>승차 위치와 하차 위치 정보를 사용해 거리를 계산하고 새로운 특징 변수를 추가</u>한다.
* 운행 거리는 유클리드 거리로 계산한다.

유클리드공식.png

위치관련 변수(거리) 만들기: 새로운 feature

승차정보와 하차정보를 좌표(경도X축,위도Y축)로 알고 있음

--> 2개의 좌표로 거리 구할 수 있음

```python
# (dropoff_longitude, dropoff_latitue)와 (pickup_longitude, pickup_lattitude)사이의 거리를 유클리드 공식으로 계산한다.
# 결과를 df['distance']에 대입한다.

df['distance'] = ((df['pickup_latitude'] - df['dropoff_latitude'])**2 + (df['pickup_longitude'] - df['dropoff_longitude'])**2)**0.5
#df['distance']: 택시가 운행한 거리정보 새로 만듦(변수,컬럼)
#[외](전체)**0.5: 루트(2분의1 제곱)
#[내](승차경도-하차경도)**2 + (승차위도-하차위도)**2: x좌표(경도) 차이에 제곱, y좌표(위도) 차이에 제곱
```

* 거리와 요금사이에는 비례관계가 있음을 알 수 있다.

* 앞에서 공항 근처에서 승차한 기록들이 있었고 공항을 오가는 택시는 52달러로 고정된 금액에 통행료를 더한 요금을 받는다. 이 정액 요금이 수직선이 생긴 이유일 것이다. 
* 따라서 <u>뉴욕시 주요 공항과 승차 및 하차 위치 사이의 거리를 새로운 특징 변수로 추가</u>할 필요가 있다. 
* 신경망은 이 특징 변수를 사용해 공항을 오고가는 정액 요금을 학습할 수 있다.

```python
# 주요 공항의 경도(longitude)와 위도(latitude)

airports = {'JFK_Airport': (-73.78,40.643),
            'Laguardia_Airport': (-73.87, 40.77),
            'Newark_Airport' : (-74.18, 40.69)}   #공항과 뉴욕시 사이를 오가는 거리에 대한 새로운 정보 필요--> 공항 딕셔너리{'키'값: 데이터(value)}
```

```python
# 경도와 위도를 사용해 거리를 계산하는 함수(euc_distance)를 정의

def euc_distance(lat1, long1, lat2, long2): #거리 구하는 공식 함수(위도x,경도y의 좌표값 두개)
    return ((lat1-lat2)**2 + (long1-long2)**2)**0.5 
```

* 승객의 승차위치부터 공항까지의 거리, 승객의 하차위치부터 공항까지의 거리를 계산한다.
* 위에 정의한 euc_distance()함수를 이용하여 세 공항과 승하차 위치사이의 거리를 계산한다.

```python
# for문을 이용하여 airports 딕셔너리에 있는 세공항과 승하차 위치사이의 거리를 계산한다.
# 'pickup_dist_' + airport = euc_distance()호출: ('pickup_latitude', airports[airport][1])과 ('pickup_longitude', airports[airport][0]) 사이의 거리
# 'dropoff_dist_' + airport = euc_distance()호출: ('dropoff_latitude', airports[airport][1])과 ('dropoff_longitude', airports[airport][0]) 사이의 거리

for airport in airports: #airports딕셔너리(3개)에서 하나씩 끄집어내어 airport(공항)에 순서대로 넣기
    df['pickup_dist_' + airport] = euc_distance(df['pickup_latitude'], df['pickup_longitude'], airports[airport][1], airports[airport][0]) #픽업한 위치부터 공항까지 = 픽업위치정보,공항위치정보 줘서 거리구하기
    df['dropoff_dist_' + airport] = euc_distance(df['dropoff_latitude'], df['dropoff_longitude'], airports[airport][1], airports[airport][0]) #드랍한 위치부터 공항까지

#feature engineering: 새로운 컬럼(feature) 만들기
#요금과 거리 사이의 +상관관계, but 거리에 상관없이 고정된 부분에는 airport(공항)에 대한 정보가 들어가 있을 것
#승객의 승차/하차 위치로부터 각 공항까지의 거리 정보 넣어주기(새로운 컬럼)--> 공항 정보로 52달러에 해당하는 부분과 거리의 상관관계 찾아낼 수 있을 것
```
